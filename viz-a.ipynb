```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load the Dataset using Seaborn

```python
mpg = sns.load_dataset('mpg')
mpg_eu = mpg[mpg['origin']=='europe']
mpg_jp = mpg[mpg['origin']=='japan']
mpg_usa = mpg[mpg['origin']=='usa']
mpg
```

# Set the data theme

```python
sns.set()
```

# The correlation between MPG vs cars Horsepower

```python
sns.barplot(x='mpg', y='horsepower', color='blue', data=mpg)
```

# The correlation between MPG vs cars Horsepower in Europe

```python
sns.lineplot(x='mpg', y='horsepower' ,color='red', data=mpg_eu)
```

# The correlation between MPG vs cars Horsepower in Japan

```python
sns.lineplot(x='mpg', y='horsepower' ,color='green', data=mpg_jp)
```

# The correlation between MPG vs cars Horsepower in USA

```python
sns.lineplot(x='mpg', y='horsepower' ,color='orange', data=mpg_usa)
```

# Conclusions:

1- Whenever the car horsepower increase the (MPG) which represent the the fuel consumption will also increase.

2- Japan have almost the highest (MPG) fuel consamption in comparison with the horsepower among the others.

3- Europe cars have the lowest horsepower and USA cars have the highest horsepower.
